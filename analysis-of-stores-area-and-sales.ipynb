```python
df = pd.read_csv('/kaggle/input/stores-area-and-sales-data/Stores.csv')
df
```

# Data visualization

There are 4 different features so there are 6 different graphs to see.

## Is the store area correlated with the number of items available ?

```python
sns.jointplot(x='Store_Area', y='Items_Available', data=df, kind="reg", truncate=False, color="m", height=7);
```

As we could expect, the bigger a shop is, the more there are items in it.

## Is the store area linked to the number of daily customers ?

We can think that big shops attract more customers than other ones, lets check this !

```python
sns.jointplot(x='Store_Area', y='Daily_Customer_Count', data=df, kind="reg", truncate=False, color="m", height=7);
```

The store area doesn't seems to have an impact on the number of daily customers.

## Do you spend more in a big store?

```python
sns.jointplot(x='Store_Area', y='Store_Sales', data=df, kind="reg", truncate=False, color="m", height=7);
```

The answer is counterintuitive since we often think we do more big shopping in big stores, but it seems that is not the case.

## The more available items the more daily customers ?

There are usually more people in big shops with a lot of different items, can we see this correlation here ?

```python
sns.jointplot(x='Items_Available', y='Daily_Customer_Count', data=df, kind="reg", truncate=False, color="m", height=7);
```

There is finally not true !

## More items available, bigger invoices ?


The question is legitimate, is the answer intuitive?

```python
sns.jointplot(x='Items_Available', y='Store_Sales', data=df, kind="reg", truncate=False, color="m", height=7);
```

There is no visible impact.

## The number of daily customers has an impact on the store sales ?

```python
sns.jointplot(x='Daily_Customer_Count', y='Store_Sales', data=df, kind="reg", truncate=False, color="m", height=7);
```

There isn't any impact on it.
